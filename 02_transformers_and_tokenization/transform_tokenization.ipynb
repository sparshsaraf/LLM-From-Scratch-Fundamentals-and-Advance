```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch 
tokenizer = AutoTokenizer.from_pretrained("gpt2")
tokenizer.padding_side = "left"
tokenizer.pad_token = tokenizer.eos_token

model = AutoModelForCausalLM.from_pretrained("gpt2")
inputs = tokenizer(
    ["Ai is good", "Healthcare in ai is"],
    return_tensors = "pt",
    padding = True
)
output = model.generate(
    inputs["input_ids"],
    attention_mask = inputs["attention_mask"],
    max_new_tokens = 50, 
    
)
for i in output:
    print(tokenizer.decode(i, skip_special_tokens = True))
    print("---")
```

```python
inputs = tokenizer("The bank by the river", return_tensors = "pt")

with torch.no_grad():
    outputs = model(**inputs, output_hidden_states = True)

hidden_states = outputs.hidden_states
print(len(hidden_states))
print(hidden_states[0].shape) 
print(hidden_states)
```

```python
last_hidden_state = hidden_states[-1]
sentence_embedding = last_hidden_state[0, -1, :]
print(sentence_embedding.shape)
print(last_hidden_state.shape)
```

```python
def get_embeddings(text):
    inputs = tokenizer(text, return_tensors = "pt")
    with torch.no_grad():
        outputs = model(**inputs, output_hidden_states = True)
    return outputs.hidden_states[-1][0, -1 , :] 

import torch.nn.functional as F

emb1 = get_embeddings("the bank by the river")
emb2 = get_embeddings("the financial bank downtown")
emb3 = get_embeddings("water flows near the riverside")

sim_12 = F.cosine_similarity(emb1.unsqueeze(0), emb2.unsqueeze(0))
sim_13 = F.cosine_similarity(emb1.unsqueeze(0), emb3.unsqueeze(0))

print(f"River bank vs financial bank: {sim_12.item():.3f}")
print(f"river bank vs riverside: {sim_13.item():.3f}")
```

```python
from sentence_transformers import SentenceTransformer

model_st = SentenceTransformer("all-MiniLM-L6-v2")

emb1 = model_st.encode("The bank by the river")
emb2 = model_st.encode("The financial bank downtown")
emb3 = model_st.encode("water flows near teh riverside")

sim_12 = F.cosine_similarity(
    torch.tensor(emb1).unsqueeze(0),
    torch.tensor(emb2).unsqueeze(0)
)
sim_13 = F.cosine_similarity(
    torch.tensor(emb1).unsqueeze(0),
    torch.tensor(emb3).unsqueeze(0)
)
print(f"River bank vs Financial bank: {sim_12.item():.3f}")
print(f"River bank vs Riverside: {sim_13.item():.3f}")
```
