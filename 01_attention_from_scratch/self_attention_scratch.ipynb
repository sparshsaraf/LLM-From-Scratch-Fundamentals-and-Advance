```python
import numpy as np 

np.random.seed(42)

#words = ["plane", "the","crashed", "did", "not" ]

embeddings = np.random.randn(5,8)

print(embeddings)
```

```python
Wq = np.random.randn(8,8)
Wk = np.random.randn(8,8)
Wv = np.random.randn(8,8)

```

```python
Q = embeddings @ Wq 
K = embeddings @ Wk 
V = embeddings @ Wv 

score = Q @ K.T 
print(score)
```

```python
updated_scores = score / (8 ** 0.5 ) 
print(updated_scores)
```

```python
def softmax(x):
    e = np.exp(x)
    
    eq = e.sum(axis = 1, keepdims = True)
    full_eq = e / eq
    return full_eq 
```

```python
attention_weights = softmax(updated_scores)
print(attention_weights)
```
